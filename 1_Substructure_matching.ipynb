```python
from rdkit import Chem
Chem
```

```python
mols=Chem.SDMolSupplier("/substruct_dataset.sdf")
mols=[mol for mol in mols if mol]
len(mols)
```

```python
smarts="n1ccccc1-[N,O]"
query_mol=Chem.MolFromSmarts(smarts)
query_mol
```

```python
matched_mols=[]
matched_mols_ats_ids = []
for mol in mols:
  at_ids= mol.GetSubstructMatch(query_mol)
  if len(at_ids):
    matched_mols.append(mol)
    matched_mols_ats_ids.append(at_ids)
```
